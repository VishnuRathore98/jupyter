```python
import mahotas as mh
from matplotlib import pyplot as plt
import numpy as np
```

```python
image = mh.imread('building_clear_sky.jpeg')
```

```python
image = image - image.mean()
```

```python
plt.imshow(image)
```

```python
plt.show()
```

```python
image=mh.colors.rgb2gray(image, dtype=np.uint8)
```

```python
plt.imshow(image)
```

```python
plt.gray()
```

```python
thresh=mh.thresholding.otsu(image)
```

```python
print(thresh)
```

```python
plt.imshow(image<thresh)
```

```python
ostubin = (image<=thresh)
```

```python
ostubin = mh.close(ostubin, np.ones((15,15)))
```
